```python
df = pd.read_csv(r'customer_shopping_behavior.csv')
df.head()
df_copy = df.copy()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
df_copy['Review Rating'] = df_copy.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df_copy.columns= df_copy.columns.map(lambda x : x.lower().replace(" ","_").strip())
```

```python
df_copy.columns
df_copy.head()
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.head()
```

```python
# create a new column age_group

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```
